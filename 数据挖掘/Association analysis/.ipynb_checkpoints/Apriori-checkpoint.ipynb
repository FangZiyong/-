```python
import pandas as pd
from IPython.display import display
from mlxtend.preprocessing import TransactionEncoder
dataset = [['Milk', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Dill', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Milk', 'Apple', 'Kidney Beans', 'Eggs'],
           ['Milk', 'Unicorn', 'Corn', 'Kidney Beans', 'Yogurt'],
           ['Corn', 'Onion', 'Onion', 'Kidney Beans', 'Ice cream', 'Eggs']]
te=TransactionEncoder()
data=te.fit(dataset).transform(dataset,sparse=True)
df=pd.SparseDataFrame(data,columns=te.columns_,default_fill_value=False)
display(df)
```

```python
####频繁集的确定from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import apriori
frequent_items=apriori(df, min_support=0.6, use_colnames=True)
frequent_items['len']=frequent_items['itemsets'].apply(lambda x:len(x))
display(frequent_items)
```

```python
from mlxtend.frequent_patterns import association_rules
rules=association_rules(frequent_items)
display(rules)
```
