# 微信朋友数据分析

itchat是个python库，利用微信网页版登陆微信，并且能够提供包括收发消息，图片，文件的功能，并能够获取微信朋友基本信息，关注的公众号等信息。这里主要使用itchat获取微信朋友的信息。

```python
##登导入itchat库并登录
import itchat
itchat.auto_login(hotReload=True)

##导入pandas库并忽略警告
import pandas as pd
pd.options.mode.chained_assignment = None

# 获取好友列表
friends = pd.DataFrame(itchat.get_friends(update=True))
```

好友列表中共包含较多字段的信息，部分为无效字段，挑选出部分字段介绍。

|    字段     | 简介                     |
| :---------: | :----------------------- |
|  UserName   | 用户名，用户的身份标识。 |
|    City     | 用户填写的城市名称。     |
|  Province   | 用户填写的省份。         |
| RemarkName  | 备注姓名。               |
|  Signature  | 用户签名。               |
|  NickName   | 用户昵称。               |
|     Sex     | 用户填写的性别。         |
| AttrStatus  | 含义未知。但可能有作用。 |
|   SnsFlag   | 含义未知。但可能有作用。 |
| ContactFlag | 含义未知。但可能有作用。 |

```python
from pyecharts import options as opts
import pyecharts
from pyecharts.charts import Pie, Map


def map(x, y, name=None, title=None) -> Map:
    c = (Map(init_opts=opts.InitOpts(width='900px', height='800px')).add(
        name, [list(z) for z in zip(list(x), list(y))],
        "china").set_global_opts(
            title_opts=opts.TitleOpts(title=title),
            visualmap_opts=opts.VisualMapOpts(max_=max(y)),
        ))
    return c.render_notebook()

def pie(x, y, title=None) -> Pie:
    c = (Pie().add(
        "", [list(z) for z in zip(list(x), list(y))]).set_global_opts(
            title_opts=opts.TitleOpts(title=title),
            legend_opts=pyecharts.options.LegendOpts(
                is_show=True,
                orient='vertical',
                pos_left='80%',
                pos_bottom='50%')).set_series_opts(label_opts=opts.LabelOpts(
                    formatter="{b}: {c}")))
    return c.render_notebook()

temp = friends['Province']
temp[temp == ''] = '未填写'
temp = temp.value_counts()
map(temp.index, temp, title='微信朋友省份分布图')
```

```python
temp=friends['Sex']
temp[temp==0]='未透露'
temp[temp==1]='男性'
temp[temp==2]='女性'
temp=temp.value_counts()
pie(temp.index,temp,title='微信朋友圈性别分布')
```
